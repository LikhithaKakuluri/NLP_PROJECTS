# Importing the Dependencies

```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Difflib

Difflib is a built-in module in the Python programming language consisting of different simple functions and classes that allow users to compare data sets. 

Cosine Similarity

cosine similarity metric measures the cosine of the angle between two n-dimensional vectors projected in a multi-dimensional space. The Cosine similarity of two documents will range from 0 to 1. If the Cosine similarity score is 1, it means two vectors have the same orientation. The value closer to 0 indicates that the two documents have less similarity.

# Data Collection and Processing

```python
df=pd.read_csv("movies.csv")
df.head()
```

```python
#number of rows and columns
df.shape
```

```python
#select the relevant features for recommendations
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with null string
for features in selected_features:
    df[features]=df[features].fillna('')
```

```python
#combining all the 5 selected features
combined_features=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
print(combined_features)
```

```python
#converting the text data to feature vectors
vectorizer=TfidfVectorizer()

feature_vectors=vectorizer.fit_transform(combined_features)

print(feature_vectors)
```

# Cosine Similarity

```python
#getting the similarity score using cosine similarity
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
#getting the movie name from the user
movie_name=input("Enter your favourite movie name:")
```

```python
#creating a list with all the movie names given in the dataset
list_of_all_titles=df['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title

index_of_the_movie=df[df.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

# Enumerate

Python's enumerate() lets you write Pythonic for loops when you need a count and the value from an iterable. The big advantage of enumerate() is that it returns a tuple with the counter and value, so you don't have to increment the counter yourself.

```python
#getting a list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
#sorting the movies based on their similarity score
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index
print("Movies suggested for you :\n")
i=1

for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if (i<30):
        print(i,'.',title_from_index)
        i+=1
```
